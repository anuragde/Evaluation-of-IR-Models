```python
import numpy as np
from sklearn.decomposition import TruncatedSVD
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
a = np.loadtxt("pca_a.txt", usecols = (0,1,2,3))
b = np.loadtxt("pca_b.txt", usecols = (0,1,2,3))
c = np.loadtxt("pca_c.txt", usecols = (0,1,2,3))
```

```python
svd1 = TruncatedSVD(n_components=2, n_iter=7, random_state=42).fit_transform(a)
svd2 = TruncatedSVD(n_components=2, n_iter=7, random_state=42).fit_transform(b)
svd3 = TruncatedSVD(n_components=2, n_iter=7, random_state=42).fit_transform(c)
```

```python
tsne1 = TSNE(learning_rate=100).fit_transform(a)
tsne2 = TSNE(learning_rate=100).fit_transform(b)
tsne3 = TSNE(learning_rate=100).fit_transform(c)
```

# Load Dataset and Visualise

```python
svd1=pd.DataFrame(svd1)
svd2=pd.DataFrame(svd2)
svd3=pd.DataFrame(svd3)
tsne1=pd.DataFrame(tsne1)
tsne2=pd.DataFrame(tsne2)
tsne3=pd.DataFrame(tsne3)
```
